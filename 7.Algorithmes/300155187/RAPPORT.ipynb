## 1ï¸âƒ£ Factorielle

### ðŸ”¹ Formule mathÃ©matique

$
n! = \begin{cases}
   1, & \text{si } n = 0 \\
   n \times (n - 1)!, & \text{si } n > 0
\end{cases}
$

```python
def factorielle_iterative(n):
    resultat = 1
    for i in range(1, n + 1):
        resultat *= i
    return resultat

# Exemple
factorielle_iterative(5)
```

```python
def factorielle_recursive(n):
    if n == 0:
        return 1
    return n * factorielle_recursive(n - 1)

# Exemple
factorielle_recursive(5)
```

```python
from functools import reduce

def factorielle_comprehension(n):
    if n == 0:  # condition de sortie (cas de base)
        return 1
    # Produit de tous les entiers de 1 Ã  n en utilisant une comprehension
    return reduce(lambda x, y: x*y, [i for i in range(1, n+1)])
```

## 2ï¸âƒ£ Somme des n premiers entiers

### ðŸ”¹ Formule mathÃ©matique

$
S(n) = 1 + 2 + 3 + \dots + n = \frac{n(n + 1)}{2}
$

### ðŸ”¹ Forme rÃ©cursive

$
S(n) =
\begin{cases}
   0, & \text{si } n = 0 \\
   n + S(n - 1), & \text{si } n > 0
\end{cases}
$

```python
def somme_while(n):
    i = 1
    s = 0
    while i <= n:
        s += i
        i += 1
    return s

somme_while(5)
```

```python
def somme_recursive(n):
    if n == 0:
        return 0
    return n + somme_recursive(n - 1)

somme_recursive(5)
```

## 3ï¸âƒ£ Suite de Fibonacci

### ðŸ”¹ Formule mathÃ©matique

$
F(n) =
\begin{cases}
   0, & \text{si } n = 0 \\
   1, & \text{si } n = 1 \\
   F(n - 1) + F(n - 2), & \text{si } n \ge 2
\end{cases}
$

```python
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

fib_iter(10)
```

```python
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n - 1) + fib_rec(n - 2)

fib_rec(10)
```
