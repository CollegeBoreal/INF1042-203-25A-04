## 1ï¸âƒ£ Factorielle

### ðŸ”¹ Formule mathÃ©matique

$
n! = \begin{cases}
   1, & \text{si } n = 0 \\
   n \times (n - 1)!, & \text{si } n > 0
\end{cases}
$

```python
def factorielle_iterative(n):
    resultat = 1
    for i in range(1, n + 1):
        resultat *= i
    return resultat

# Exemple
factorielle_iterative(5)
```

```python
def factorielle_recursive(n):
    if n == 0:
        return 1
    return n * factorielle_recursive(n - 1)

# Exemple
factorielle_recursive(5)
```

```python
from functools import reduce

def factorielle_comprehension(n):
    if n == 0:  # condition de sortie (cas de base)
        return 1
    # Produit de tous les entiers de 1 Ã  n en utilisant une comprehension
    return reduce(lambda x, y: x*y, [i for i in range(1, n+1)])
```

#### ðŸ“ SynthÃ¨se (Factorielle)

Les trois versions produisent le mÃªme rÃ©sultat. La version **itÃ©rative** est la plus efficace en Python standard, car elle Ã©vite la surcharge des appels de fonction successifs (Stack Overflow). La version **rÃ©cursive** est la traduction la plus directe de la formule mathÃ©matique. La version par **comprÃ©hension** et `reduce` est une alternative plus fonctionnelle et concise.

## 2ï¸âƒ£ Somme des n premiers entiers

### ðŸ”¹ Formule mathÃ©matique

$$
S(n) = 1 + 2 + 3 + \dots + n = \frac{n(n + 1)}{2}
$$

### ðŸ”¹ Forme rÃ©cursive

$$
S(n) =
\begin{cases}
   0, & \text{si } n = 0 \\
   n + S(n - 1), & \text{si } n > 0
\end{cases}
$$

```python
def somme_while(n):
    i = 1
    s = 0
    while i <= n:
        s += i
        i += 1
    return s

# Exemple
somme_while(5)
```

```python
def somme_recursive(n):
    if n == 0:
        return 0
    return n + somme_recursive(n - 1)

# Exemple
somme_recursive(5)
```

#### ðŸ“ SynthÃ¨se (Somme)

La fonction de somme illustre bien le parallÃ¨le entre la boucle **`while` (itÃ©rative)** et l'appel rÃ©cursif. La version rÃ©cursive s'arrÃªte lorsque le **cas de base** ($n=0$) est atteint, tandis que la version itÃ©rative s'arrÃªte lorsque la **condition de la boucle** ($i \le n$) est fausse. Les deux ont une complexitÃ© temporelle similaire, $O(n)$.

## 3ï¸âƒ£ Suite de Fibonacci

### ðŸ”¹ Formule mathÃ©matique

$$
F(n) =
\begin{cases}
   0, & \text{si } n = 0 \\
   1, & \text{si } n = 1 \\
   F(n - 1) + F(n - 2), & \text{si } n \ge 2
\end{cases}
$$

```python
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

# Exemple
fib_iter(10)
```

```python
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n - 1) + fib_rec(n - 2)

# Exemple
fib_rec(10)
```

#### ðŸ“ SynthÃ¨se (Fibonacci)

L'algorithme de Fibonacci est un cas d'Ã©tude important :
- La version **itÃ©rative** est trÃ¨s rapide, $O(n)$, car elle calcule chaque terme une seule fois.
- La version **rÃ©cursive** est extrÃªmement lente, $O(2^n)$, pour les grands $n$, car elle refait des calculs (par exemple, $F(5)$ est calculÃ© plusieurs fois). Cela dÃ©montre l'importance d'analyser la **complexitÃ©** de l'algorithme.
