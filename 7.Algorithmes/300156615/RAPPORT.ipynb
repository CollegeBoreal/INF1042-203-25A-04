## 1ï¸âƒ£ Factorielle

### ðŸ”¹ Formule mathÃ©matique

[
n! =
\begin{cases}
1, & \text{si } n = 0 \
n \times (n - 1)!, & \text{si } n > 0
\end{cases}
]

### ðŸŸ§ Version itÃ©rative

```python
def factorielle_iterative(n):
    resultat = 1
    for i in range(1, n + 1):
        resultat *= i
    return resultat

factorielle_iterative(5)
```

### ðŸŸ§ Version rÃ©cursive

```python
def factorielle_recursive(n):
    if n == 0:
        return 1
    return n * factorielle_recursive(n - 1)

factorielle_recursive(5)
```

### ðŸŸ§ Version fonctionnelle (reduce + comprehension)

```python
from functools import reduce

def factorielle_comprehension(n):
    if n == 0:
        return 1
    return reduce(lambda x, y: x * y, [i for i in range(1, n + 1)])

factorielle_comprehension(5)
```

## 2ï¸âƒ£ Somme des n premiers entiers

### ðŸ”¹ Formule mathÃ©matique

[
S(n) = 1 + 2 + ... + n = \frac{n(n + 1)}{2}
]

### ðŸ”¹ Forme rÃ©cursive

[
S(n) =
\begin{cases}
0, & \text{si } n = 0 \
n + S(n - 1), & \text{si } n > 0
\end{cases}
]

### ðŸŸ§ Version itÃ©rative (while)

```python
def somme_while(n):
    i = 1
    s = 0
    while i <= n:
        s += i
        i += 1
    return s

somme_while(5)
```

### ðŸŸ§ Version rÃ©cursive

```python
def somme_recursive(n):
    if n == 0:
        return 0
    return n + somme_recursive(n - 1)

somme_recursive(5)
```

## 3ï¸âƒ£ Suite de Fibonacci

### ðŸ”¹ Formule mathÃ©matique

[
F(n) =
\begin{cases}
0, & n = 0 \
1, & n = 1 \
F(n - 1) + F(n - 2), & n \ge 2
\end{cases}
]

### ðŸŸ§ Version itÃ©rative

```python
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

fib_iter(10)
```

### ðŸŸ§ Version rÃ©cursive

```python
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n - 1) + fib_rec(n - 2)

fib_rec(10)
```
