```python
import matplotlib.pyplot as plt

# DÃ©finition des fonctions
def factorielle_iterative(n):
    resultat = 1
    for i in range(1, n + 1):
        resultat *= i
    return resultat

def somme_while(n):
    i = 1
    s = 0
    while i <= n:
        s += i
        i += 1
    return s

def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

# DonnÃ©es
n = list(range(10))
fact = [factorielle_iterative(i) for i in n]
somme = [somme_while(i) for i in n]
fib = [fib_iter(i) for i in n]

# Visualisation
plt.figure(figsize=(10,6))
plt.plot(n, fact, label="Factorielle", marker="o")
plt.plot(n, somme, label="Somme", marker="s")
plt.plot(n, fib, label="Fibonacci", marker="^")

plt.title("Comparaison de la croissance des algorithmes")
plt.xlabel("n")
plt.ylabel("Valeurs calculÃ©es")
plt.legend()
plt.grid(True)
plt.show()
```
