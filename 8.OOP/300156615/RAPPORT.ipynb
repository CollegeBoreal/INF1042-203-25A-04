```python
from carre import Carre
from cercle import Cercle

c1 = Carre(4)
c2 = Cercle(3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"CarrÃ© â€” cÃ´tÃ©={cote}, aire={carre.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle â€” rayon={r}, aire={cercle.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
afficher_carre(c1)
afficher_cercle(c2)
```

```python
def afficher_triangle(triangle):
    base = triangle.base
    h = triangle.hauteur

    # CoordonnÃ©es des 3 sommets
    x = [0, base, base/2, 0]
    y = [0, 0, h, 0]

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Triangle â€” base={base}, hauteur={h}, aire={triangle.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
def afficher_rectangle(rect):
    L = rect.largeur
    H = rect.hauteur

    x = [0, L, L, 0, 0]
    y = [0, 0, H, H, 0]

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Rectangle â€” largeur={L}, hauteur={H}, aire={rect.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
from Triangle import Triangle
from Rectangle import Rectangle

t1 = Triangle(10, 6)
r1 = Rectangle(12, 7)

print(t1.afficher_info())
print(r1.afficher_info())

afficher_triangle(t1)
afficher_rectangle(r1)
```

```python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from Parallelepipede3D import Parallelepipede3D
import matplotlib.pyplot as plt
import numpy as np

def afficher_parallelepipede3D(p):
    S = p.sommets()  # 8 sommets

    faces_idx = [
        [0, 1, 3, 2],  
        [4, 5, 7, 6],  
        [0, 1, 5, 4],
        [2, 3, 7, 6],
        [0, 2, 6, 4],
        [1, 3, 7, 5],
    ]

    faces = [S[idx] for idx in faces_idx]

    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection="3d")

    poly = Poly3DCollection(faces, alpha=0.3, edgecolor="k")
    ax.add_collection3d(poly)

    xs, ys, zs = S[:, 0], S[:, 1], S[:, 2]
    max_range = max(xs.max()-xs.min(), ys.max()-ys.min(), zs.max()-zs.min()) / 2

    mid_x = (xs.max() + xs.min()) / 2
    mid_y = (ys.max() + ys.min()) / 2
    mid_z = (zs.max() + zs.min()) / 2

    ax.set_xlim(mid_x - max_range, mid_x + max_range)
    ax.set_ylim(mid_y - max_range, mid_y + max_range)
    ax.set_zlim(mid_z - max_range, mid_z + max_range)

    ax.set_title(f"ParallÃ©lÃ©pipÃ¨de 3D â€” volume = {p.volume():.2f}")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

    plt.grid(True)
    plt.show()
```

```python
p = Parallelepipede3D(
    origine=[0, 0, 0],
    v1=[3, 0, 0],
    v2=[1, 2, 0],
    v3=[0, 0, 4]
)

print("Volume :", p.volume())
afficher_parallelepipede3D(p)
```
