```python
from Carre import Carre
from Cercle import Cercle
from Trapeze import Trapeze


c1 = Carre(4)
c2 = Cercle(3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def afficher_Carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"CarrÃ© â€” cÃ´tÃ©={cote}, aire={carre.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
def afficher_Cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle â€” rayon={r}, aire={cercle.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python


def afficher_Trapeze(grande_base, petite_base, hauteur):
   
    A = (0, 0)                    
    B = (grande_base, 0)           

   
    offset = (grande_base - petite_base) / 2

    D = (offset, hauteur)          # Haut gauche
    C = (offset + petite_base, hauteur)  # Haut droit

  
    X = [A[0], B[0], C[0], D[0], A[0]]
    Y = [A[1], B[1], C[1], D[1], A[1]]

   
    aire = (grande_base + petite_base) * hauteur / 2

  
    plt.figure(figsize=(6, 6))
    plt.plot(X, Y, 'b-')
    plt.fill(X, Y, alpha=0.3)
    plt.title(f"TrapÃ¨ze â€” B={grande_base}, b={petite_base}, h={hauteur},  aire={aire:.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()

```

```python
afficher_Carre(c1)
afficher_Cercle(c2)
afficher_Trapeze(20, 10, 5)
```
