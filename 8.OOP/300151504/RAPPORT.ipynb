## ‚ñ∂Ô∏è Importer les classes

```python
from Carre import Carre
from Cercle import Cercle
from Triangle import Triangle

c1 = Carre(4)
c2 = Cercle(3)
t1 = Triangle(6,4)

print(c1.afficher_info())
print(c2.afficher_info())
print(t1.afficher_info())
```

## üé® Import pour affichage graphique

```python
import numpy as np
import matplotlib.pyplot as plt
```

## üü¶ Afficher Carr√©

```python
def afficher_carre(carre: Carre):
    x = [0, carre.cote, carre.cote, 0, 0]
    y = [0, 0, carre.cote, carre.cote, 0]
    plt.figure(figsize=(5,5))
    plt.plot(x,y)
    plt.fill(x,y,alpha=0.3)
    plt.title(f"Carr√© ‚Äî c√¥t√©={carre.cote}, aire={carre.aire()}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

## ‚ö™ Afficher Cercle

```python
def afficher_cercle(cercle: Cercle):
    theta = np.linspace(0,2*np.pi,300)
    x = cercle.rayon * np.cos(theta)
    y = cercle.rayon * np.sin(theta)
    plt.figure(figsize=(5,5))
    plt.plot(x,y)
    plt.fill(x,y,alpha=0.3)
    plt.title(f"Cercle ‚Äî rayon={cercle.rayon}, aire={cercle.aire():.2f}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

## üî∫ Afficher Triangle

```python
def afficher_triangle(triangle: Triangle):
    x = [0, triangle.base, triangle.base/2, 0]
    y = [0, 0, triangle.hauteur, 0]
    plt.figure(figsize=(5,5))
    plt.plot(x,y,marker='o')
    plt.fill(x,y,alpha=0.3)
    plt.title(f"Triangle ‚Äî base={triangle.base}, hauteur={triangle.hauteur}, aire={triangle.aire():.2f}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

## üñºÔ∏è Affichage final

```python
afficher_carre(c1)
afficher_cercle(c2)
afficher_triangle(t1)
```
