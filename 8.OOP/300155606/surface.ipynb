```python
# Calcul de la surface d'un trapèze avec input

grande_base = float(input("Entrez la grande base (B) : "))
petite_base = float(input("Entrez la petite base (b) : "))
hauteur = float(input("Entrez la hauteur (h) : "))

surface = ((grande_base + petite_base) * hauteur) / 2

print("La surface du trapèze est :", surface)
```
