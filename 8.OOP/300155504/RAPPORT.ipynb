```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Carré - côté={cote}, aire={carre.aire()}")

    # حفظ الصورة
    plt.savefig("carre.png")
    plt.show()


def afficher_cercle(cercle):
    rayon = cercle.rayon
    theta = np.linspace(0, 2 * np.pi, 200)

    x = rayon * np.cos(theta)
    y = rayon * np.sin(theta)

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle - rayon={rayon}, aire={cercle.aire():.2f}")

    # حفظ الصورة
    plt.savefig("cercle.png")
    plt.show()
```
