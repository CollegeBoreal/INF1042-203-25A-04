# Notebook : Figures g√©om√©triques en Python
Ce notebook teste les classes :
- `Carre`
- `Cercle`
- `Trapeze`
- `Hexagone`

Nous afficherons les aires et une repr√©sentation graphique simple.


```python
from carre import Carre
from cercle import Cercle
from trapeze import Trapeze
from hexagone import Hexagone

import matplotlib.pyplot as plt
import numpy as np
```

# Test du carr√©
Cr√©ons un carr√© et affichons ses informations.


```python
c = Carre(4)
print(c.afficher_info())
```

## Affichage graphique du carr√©
Nous tra√ßons un carr√© simple avec matplotlib.


```python
cote = c.cote
x = [0, cote, cote, 0, 0]
y = [0, 0, cote, cote, 0]

plt.plot(x, y)
plt.title("Carr√©")
plt.axis("equal")
plt.show()
```

# Test du cercle
Cr√©ons un cercle et affichons ses informations.


```python
cer = Cercle(3)
print(cer.afficher_info())
```

## Affichage graphique du cercle
Nous tra√ßons un cercle avec une approximation param√©trique.


```python
r = cer.rayon
theta = np.linspace(0, 2*np.pi, 200)
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(x, y)
plt.title("Cercle")
plt.axis("equal")
plt.show()
```

# Test du trap√®ze


```python
t = Trapeze(5, 7, 4)   # base1, base2, hauteur
print(t.afficher_info())
```

## Affichage graphique du trap√®ze
Nous tra√ßons un trap√®ze d√©fini par :
- base1 (bas)
- base2 (haut)
- hauteur

Le trap√®ze est centr√© horizontalement pour une meilleure visualisation.


```python
# R√©cup√©ration des mesures
b1 = t.base1   # grande base
b2 = t.base2   # petite base
h = t.hauteur  # hauteur

# Pour que le trap√®ze soit centr√©, on d√©cale la petite base
decalage = (b1 - b2) / 2

# Coordonn√©es du trap√®ze
x = [0, b1, b2 + decalage, decalage, 0]
y = [0, 0, h, h, 0]

plt.plot(x, y)
plt.title("Trap√®ze")
plt.axis("equal")
plt.show()
```

# Test de l‚Äôhexagone r√©gulier


```python
h = Hexagone(4)
print(h.afficher_info())
```

```python
## Affichage graphique de l‚Äôhexagone
```

```python
r = h.cote
angles = np.linspace(0, 2*np.pi, 7)
x = r * np.cos(angles)
y = r * np.sin(angles)

plt.plot(x, y)
plt.title("Hexagone r√©gulier")
plt.axis("equal")
plt.show()
```
