```python
from carre import Carre
from cercle import Cercle

c1 = Carre(4)
c2 = Cercle(3)

print(c1.afficher_info())
print(c2.afficher_info())
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Carr√© ‚Äî c√¥t√©={cote}, aire={carre.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle ‚Äî rayon={r}, aire={cercle.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
afficher_carre(c1)
afficher_cercle(c2)
```

## üî∂ Trap√®ze

### Formule de l‚Äôaire

\[
A = \frac{(B + b) \times h}{2}
\]

o√π :
- **B** = grande base  
- **b** = petite base  
- **h** = hauteur  

### Exemple  
B = 10  
b = 6  
h = 4  

Calcul :  
\[
A = \frac{(10 + 6) \times 4}{2} = 32
\]

‚úÖ **Aire = 32 unit√©s¬≤**

```python
class Trapeze:
    def __init__(self, grande_base, petite_base, hauteur):
        self.B = grande_base
        self.b = petite_base
        self.h = hauteur

    def aire(self):
        return ((self.B + self.b) * self.h) / 2

    def afficher_info(self):
        return (
            f"Trap√®ze : B={self.B}, b={self.b}, h={self.h}, aire={self.aire()}"
        )
```

```python
t = Trapeze(10, 6, 4)
print(t.afficher_info())
```

## üîµ Ellipse

### Formule de l‚Äôaire

\[
A = \pi \times a \times b
\]

o√π :
- **a** = demi-grand axe  
- **b** = demi-petit axe  

### Exemple  
a = 6  
b = 4  

Calcul :  
\[
A = \pi \times 6 \times 4 = 75.39
\]

(approximation √† deux d√©cimales)

```python
import math

class Ellipse:
    def __init__(self, a, b):
        self.a = a  # demi-grand axe
        self.b = b  # demi-petit axe

    def aire(self):
        return math.pi * self.a * self.b

    def afficher_info(self):
        return f"Ellipse : a={self.a}, b={self.b}, aire={self.aire():.2f}"
```

```python
def afficher_rectangle(rect):
    L = rect.L
    l = rect.l

    x = [0, L, L, 0, 0]
    y = [0, 0, l, l, 0]

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Rectangle ‚Äî L={L}, l={l}, aire={rect.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
def afficher_ellipse(ellipse):
    a = ellipse.a
    b = ellipse.b

    theta = np.linspace(0, 2*np.pi, 300)
    x = a * np.cos(theta)
    y = b * np.sin(theta)

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Ellipse ‚Äî a={a}, b={b}, aire={ellipse.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```
