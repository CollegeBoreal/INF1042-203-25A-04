```python
from Carre import Carre
from Cercle import Cercle
from Hexagone import Hexagone
from Rectangle import Rectangle 
from Parallelogramme import Parallelogramme 

c1 = Carre(4)
c2 = Cercle(3)
c3 = Rectangle(4, 7)
c4 = Hexagone(6)
c5 = Parallelogramme(5, 3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Carr√© ‚Äî c√¥t√©={cote}, aire={carre.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle ‚Äî rayon={r}, aire={cercle.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
def afficher_hexagon(hexagone):
    angles = np.linspace(0, 2*np.pi, 7)  # 6 c√¥t√©s + retour au point de d√©part
    x = np.cos(angles)
    y = np.sin(angles)

    plt.figure(figsize=(5,5))
    plt.plot(x, y, marker='o')
    plt.title("Hexagone")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
def afficher_rectangle(rectangle):
    # Coordonn√©es du rectangle (0,0) -> (2,1)
    x = [0, 2, 2, 0, 0]
    y = [0, 0, 1, 1, 0]

    plt.figure(figsize=(5,5))
    plt.plot(x, y, marker='o')
    plt.title("Rectangle simple")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
def afficher_parallelogram(parallelogramme):
    x = [0, 3, 5, 2, 0]  # 4 sommets + retour au d√©but
    y = [0, 0, 2, 2, 0]

    plt.figure(figsize=(5,5))
    plt.plot(x, y, marker='o')
    plt.title("Parall√©logramme simple")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
afficher_carre(c1)
afficher_cercle(c2)
afficher_parallelogram(c5)
afficher_rectangle(c3)
afficher_hexagon(c4)
```
