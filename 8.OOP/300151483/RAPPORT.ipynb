```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

fig = plt.figure(figsize=(12, 10))
titles = ["Cube (CarrÃ©)", "SphÃ¨re (Cercle)", "Pyramide triangulaire (Triangle)", "Prisme trapÃ©zoÃ¯dal (TrapÃ¨ze)"]

# --- Cube ---
ax1 = fig.add_subplot(221, projection='3d')
r = [0, 4]
X, Y = np.meshgrid(r, r)
ax1.plot_surface(X, Y, np.full_like(X, 0), alpha=0.5)
ax1.plot_surface(X, Y, np.full_like(X, 4), alpha=0.5)
ax1.set_title(titles[0])
ax1.set_box_aspect([1,1,1])

# --- SphÃ¨re ---
ax2 = fig.add_subplot(222, projection='3d')
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = 3*np.cos(u)*np.sin(v)
y = 3*np.sin(u)*np.sin(v)
z = 3*np.cos(v)
ax2.plot_surface(x, y, z, color='lightblue', alpha=0.6)
ax2.set_title(titles[1])
ax2.set_box_aspect([1,1,1])

# --- Pyramide triangulaire ---
ax3 = fig.add_subplot(223, projection='3d')
vertices = [
    [0, 0, 0], [5, 0, 0], [2.5, 2, 0], [2.5, 1, 3]
]
faces = [
    [vertices[0], vertices[1], vertices[2]],
    [vertices[0], vertices[1], vertices[3]],
    [vertices[1], vertices[2], vertices[3]],
    [vertices[2], vertices[0], vertices[3]]
]
ax3.add_collection3d(Poly3DCollection(faces, facecolors='lightgreen', linewidths=1, edgecolors='black', alpha=0.7))
ax3.set_title(titles[2])
ax3.set_box_aspect([1,1,1])

# --- Prisme trapÃ©zoÃ¯dal ---
ax4 = fig.add_subplot(224, projection='3d')
base1 = [[0, 0, 0], [6, 0, 0], [5, 3, 0], [1, 3, 0]]
base2 = [[0, 0, 4], [6, 0, 4], [5, 3, 4], [1, 3, 4]]
faces = [
    base1, base2,
    [base1[0], base1[1], base2[1], base2[0]],
    [base1[1], base1[2], base2[2], base2[1]],
    [base1[2], base1[3], base2[3], base2[2]],
    [base1[3], base1[0], base2[0], base2[3]]
]
ax4.add_collection3d(Poly3DCollection(faces, facecolors='salmon', linewidths=1, edgecolors='black', alpha=0.7))
ax4.set_title(titles[3])
ax4.set_box_aspect([1,1,1])

plt.tight_layout()
plt.show()

```
