Nous allons aussi afficher graphiquement ces figures grÃ¢ce Ã  **matplotlib** et **numpy**.

```python
from Carre import Carre
from Cercle import Cercle
from Triangle import Triangle

# CrÃ©ation des objets
c1 = Carre(4)
c2 = Cercle(3)
t1 = Triangle(5, 2)

# Affichage des informations
for f in [c1, c2, t1]:
    print(f.afficher_info())
```

```python
import numpy as np
import matplotlib.pyplot as plt

def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]
    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"CarrÃ© â€” cÃ´tÃ©={cote}, aire={carre.aire()}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)
    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle â€” rayon={r}, aire={cercle.aire():.2f}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
def afficher_triangle(triangle):
    b = triangle.base
    h = triangle.hauteur
    x = [0, b, b/2, 0]
    y = [0, 0, h, 0]

    plt.figure(figsize=(5,5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Triangle â€” base={b}, hauteur={h}, aire={triangle.aire()}")
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
from Carre import Carre
from Cercle import Cercle
from Triangle import Triangle

# CrÃ©ation des objets
c1 = Carre(4)
c2 = Cercle(3)
t1 = Triangle(5, 2)

# Affichage des informations
for f in [c1, c2, t1]:
    print(f.afficher_info())
```

```python
afficher_carre(c1)
afficher_cercle(c2)
afficher_triangle(t1)
```
