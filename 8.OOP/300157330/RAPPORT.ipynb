```python
from Carre import Carre
from Cercle import Cercle
from triangle import Triangle

c1 = Carre(4)
c2 = Cercle(3)
t1 = Triangle (5,6)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def afficher_carre(carre):
    cote = carre.cote
    x = [0, cote, cote, 0, 0]
    y = [0, 0, cote, cote, 0]

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Carr√© ‚Äî c√¥t√©={cote}, aire={carre.aire()}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
def afficher_cercle(cercle):
    r = cercle.rayon
    theta = np.linspace(0, 2*np.pi, 300)

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(5, 5))
    plt.plot(x, y)
    plt.fill(x, y, alpha=0.3)
    plt.title(f"Cercle ‚Äî rayon={r}, aire={cercle.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt

def afficher_Triangle(triangle):
    """
    Affiche graphiquement un triangle d√©fini par base et hauteur.
    On suppose un triangle rectangle :
    A = (0, 0)
    B = (base, 0)
    C = (0, hauteur)
    """

    # Extraction des valeurs
    base = triangle.base
    hauteur = triangle.hauteur

    # Coordonn√©es des sommets
    x = [0, base, 0, 0]       # On revient au point A pour fermer la figure
    y = [0, 0, hauteur, 0]

    # Trac√©
    plt.figure(figsize=(5, 5))
    plt.plot(x, y)                 # contour
    plt.fill(x, y, alpha=0.3)      # remplissage
    plt.title(f"Triangle ‚Äî base={base}, hauteur={hauteur}, aire={triangle.aire():.2f}")
    plt.axis("equal")
    plt.grid(True)
    plt.show()
```

```python
afficher_carre(c1)
afficher_cercle(c2)
afficher_Triangle(t1)
```
