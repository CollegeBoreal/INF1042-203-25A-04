## 1Ô∏è‚É£ Module `embellisseur` ‚Äî Polymorphisme

### üü© Cellule Markdown

Cette partie montre comment `f` traite plusieurs types de donn√©es.

```python
from embellisseur import f

exemples = [10, 2.718, "hello", [7,8,9], (10,20,30), {"x":1, "y":2}]

for e in exemples:
    print(f"\n--- f({e}) ---")
    resultat = f(e)
    for k, v in resultat.items():
        print(f"{k:12} -> {v}")
```

## 3Ô∏è‚É£ Ensembles et compr√©hension

### üü© Markdown

Construction d‚Äôun ensemble filtr√©.

### üüß Code

```python
K = ['A', 'B', 'C']
X = [1, 2, 3, 4, 5, 6]

def P(x):
    return x % 2 == 0

S = {(k, x) for k in K for x in X if P(x)}
print("Ensemble filtr√© :", S)

x_values = [x for (_, x) in S]

moyenne = algebrique(x_values) if x_values else 0
print("Moyenne alg√©brique des x filtr√©s :", moyenne)
```
