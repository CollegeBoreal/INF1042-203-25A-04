```python
## 1Ô∏è‚É£ Import du module `embellisseur`

Cette section montre comment la fonction `f` traite diff√©rents **types** de variables.  
Elle illustre le **polymorphisme** : une m√™me fonction peut g√©rer plusieurs types.
```

```python
from embellisseur import f

exemples = [
    10,
    2.718,
    "hello",
    [7, 8, 9],
    (10, 20, 30),
    {"x": 1, "y": 2}
]

for e in exemples:
    print(f"\n--- f({e!r}) ---")
    resultat = f(e)
    for k, v in resultat.items():
        print(f"{k:18} -> {v}")
```

```python
## 2Ô∏è‚É£ Moyenne alg√©brique et g√©om√©trique

- `algebrique(L)` : moyenne arithm√©tique  
  $$\bar{x}=\frac{\sum_{i=1}^{n} x_i}{n}$$

- `geometrique(L)` : moyenne g√©om√©trique  
  $$G=\sqrt[n]{\prod_{i=1}^{n} x_i}$$
```

```python
from moyenne import algebrique, geometrique

donnees = [2, 4.5, 6, 8, 10]
print("Donn√©es :", donnees)
print("Moyenne alg√©brique :", algebrique(donnees))
print("Moyenne g√©om√©trique :", geometrique(donnees))
```

```python
## 3Ô∏è‚É£ Ensembles (set) et compr√©hension

On construit un ensemble `S` de couples `(k, x)` o√π :

- `k ‚àà K = ['A', 'B', 'C']`
- `x ‚àà X = [1, 2, 3, 4, 5, 6]`
- `x` est un nombre **pair**
```

```python
K = ['A', 'B', 'C']
X = [1, 2, 3, 4, 5, 6]

def P(x):
    return x % 2 == 0

S = {(k, x) for k in K for x in X if P(x)}
print("Ensemble des couples filtr√©s :", S)

x_values = [x for (_, x) in S]
moyenne_pairs = algebrique(x_values) if x_values else 0
print("Moyenne alg√©brique des x filtr√©s :", moyenne_pairs)
```
