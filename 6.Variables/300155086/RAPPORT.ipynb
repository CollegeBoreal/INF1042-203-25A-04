## Objectif
Ce rapport explore les notions de polymorphisme, de moyennes (algébrique et géométrique), et de compréhension d’ensembles en Python.


```python
from embellisseur import f

exemples = [10, 2.718, "hello", [7,8,9], (10,20,30), {"x":1,"y":2}]
for e in exemples:
    print(f"\n--- f({e}) ---")
    resultat = f(e)
    for k,v in resultat.items():
        print(f"{k:12} -> {v}")
```

```python
from moyenne import algebrique, geometrique

L = [2, 4.5, 6, 8, 10]
print("Moyenne algébrique :", algebrique(L))
print("Moyenne géométrique :", geometrique(L))
```

### Ensembles filtrés et compréhension

On utilise une compréhension d’ensemble pour créer des couples `(k, x)` où `x` est un entier pair.  
Ensuite, on extrait les valeurs `x` pour calculer leur moyenne algébrique.


```python
K = ['A', 'B', 'C']
X = [1, 2, 3, 4, 5, 6]

def P(x): return x % 2 == 0  # filtre les nombres pairs

S = {(k, x) for k in K for x in X if P(x)}
print("Ensemble des couples filtrés :", S)

x_values = [x for _, x in S]
from moyenne import algebrique
moyenne = algebrique(x_values) if x_values else 0
print("Moyenne des x filtrés :", moyenne)
```
