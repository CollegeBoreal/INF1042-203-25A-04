```python
ü™ê RAPPORT : Analyse de Code Python

Variables ‚Äî Moyennes ‚Äî Ensembles
| √âl√©ment | Information        |
|--------|--------------------|
| Nom    | Chouaib Aitchouaib |
| üÜî     | 300155045          |
    1Ô∏è‚É£ Analyse du module embellisseur

Cette section teste la fonction f avec diff√©rents types de donn√©es.
Elle permet d‚Äôobserver comment Python traite le polymorphisme.

üî∏ Code
from embellisseur import f

exemples = [
    10,
    2.718,
    "hello",
    [7, 8, 9],
    (10, 20, 30),
    {"x": 1, "y": 2}
]

for e in exemples:
    print(f"\n--- f({e}) ---")
    resultat = f(e)
    for k, v in resultat.items():
        print(f"{k:12} -> {v}")

2Ô∏è‚É£ Calcul des Moyennes (Alg√©brique et G√©om√©trique)

Dans cette section, nous utilisons le module moyenne qui contient :

algebrique(L) : calcule la moyenne arithm√©tique

geometrique(L) : calcule la moyenne g√©om√©trique

üî∏ Code
from moyenne import algebrique, geometrique

L = [2, 4.5, 6, 8, 10]

print("Moyenne alg√©brique :", algebrique(L))
print("Moyenne g√©om√©trique :", geometrique(L))

3Ô∏è‚É£ Ensembles (set) et Compr√©hensions

Nous construisons un ensemble S de couples (k, x) o√π :

k dans ['A', 'B', 'C']

x dans [1, 2, 3, 4, 5, 6]

x est pair

Ensuite, nous calculons la moyenne alg√©brique des valeurs retenues.

üî∏ Code
from moyenne import algebrique

K = ['A', 'B', 'C']
X = [1, 2, 3, 4, 5, 6]

def P(x):
    return x % 2 == 0

S = {(k, x) for k in K for x in X if P(x)}
print("Ensemble des couples filtr√©s :", S)

x_values = [x for (_, x) in S]
print("Valeurs filtr√©es :", x_values)

if x_values:
    moyenne = algebrique(x_values)
else:
    moyenne = 0

print("Moyenne alg√©brique des x filtr√©s :", moyenne)

üß† Conclusion

Ce rapport met en √©vidence plusieurs points importants :

Une m√™me fonction (f) peut traiter plusieurs types de donn√©es (polymorphisme).

Les fonctions algebrique et geometrique permettent d‚Äôanalyser des listes num√©riques.

Les ensembles (set) et les compr√©hensions facilitent le filtrage et l‚Äôanalyse.

Python est un langage flexible pour manipuler et transformer des donn√©es.
```
